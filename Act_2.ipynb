```python
from scipy import integrate
import scipy
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import pylab
```

```python
k = 1.3806e-23
h = 6.6262e-34
c = 3e+8

def pbr(l, t):
    a = 2.0*h*c**2
    b = (h*c) / (l*k*t)
    pbr2 = a / ((l**5)*(np.exp(b) - 1.0))
    return pbr2

wav = np.arange(350e-9, 750e-9, 1e-9)
t1 = pbr(wav, 1000)
t2 = pbr(wav, 2500)
t3 = pbr(wav, 5400)
t4 = pbr(wav, 6500)
pylab.plot(wav*1e9, (t1- min(t1))/(max(t1) - min(t1)), 'b-', label='1000K') 
pylab.plot(wav*1e9, (t2- min(t2))/(max(t2) - min(t2)), 'g-', label='2500K') 
pylab.plot(wav*1e9, (t3- min(t3))/(max(t3) - min(t3)), 'r-', label='5400K') 
pylab.plot(wav*1e9, (t4- min(t4))/(max(t4) - min(t4)), 'm-', label='6500K') 
pylab.legend(loc='upper left')
plt.style.use('ggplot')
plt.ylabel('Emmitance per Wavelength interval')
plt.xlabel('Wavelength (nm)')
plt.title("Planck's Blackbody Radiation")
plt.show()
```
